```python
import wandb
```

```python
langs = [
    "Belarusian",
    "Bulgarian",
    "Czech",
    "English",
    "Estonian",
    "Croatian",
    "Hungarian",
    "Armenian",
    "Lithuanian",
    "Macedonian",
    "Polish",
    "Romanian",
    "Russian",
    "Slovenian",
    "Serbian",
    "Ukrainian",
]
```

```python
url = "https://wandb.ai/eenlp/paraphrase_detection/runs/14zchc5r"
```

```python
run = wandb.Api().run(url.replace("https://wandb.ai/", ""))
```

```python
run.config["MODEL"]
```

```python
history = run.history()
for lang in langs:
    print(str(round(history[f"valid/acc/{lang.lower()}"].max() * 100, 2)) + "%")
```
